```python
%matplotlib inline

import pandas as pd
import numpy as np
from data_model import PerfProDataModel
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

# create the data model
perfpro_file = r'D:\perfpro_data\PerfPRO.sqlite'
pp = PerfProDataModel(perfpro_file)
```

## Leitura dos dados

```python
#user =  'Japao Carvalho'

#train_data_user = pp.get_performance_by_user(user)
train_performances = pp.get_performances()
train_performances.dateStamp = pd.to_datetime(train_performances.dateStamp)

train_performances = train_performances[train_performances.dateStamp >= '2021']

train_performances.head(10)
```

## Visualização

```python
user = 'Victor São Paulo  Ruela' # 'Japao Carvalho' # 'Flavia Camello Teixeira' #'Victor São Paulo  Ruela' 
data_user = train_performances[train_performances.userName == user]

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(data_user.dateStamp, data_user.NormPower, 'bo-', label='Potência')
ax.plot(data_user.dateStamp, data_user.NormPower.rolling(10).mean(), 'r--', label='Média Movel')
ax.plot(data_user.dateStamp, data_user.UserMaxWatts, 'kd', label='FTP')

ax.legend()
ax.set_title(user)

fig, ax = plt.subplots(figsize=(12,6))
sns.histplot(data=data_user, x='NormPower' , hue='UserMaxWatts', kde=True, palette='Set1')
ax.set_title(user)
```

```python
user_frequence = train_performances.groupby('userName')['dateStamp'].count().sort_values(ascending=False).reset_index()
user_frequence.columns = ['Atleta','Treinos']
user_frequence = user_frequence[user_frequence.Treinos > 10]

fig, ax = plt.subplots(figsize=(6,12))
sns.barplot(data=user_frequence, y='Atleta', x='Treinos', ax=ax, palette='hot')
ax.set_title('Mais Frequentes - 2021')
```

```python
user_power =  train_performances.loc[train_performances.userName.isin(user_frequence[user_frequence.Treinos > 10].Atleta), ['userName','NormPower']]# train_performances.groupby('userName')['NormPower']
user_power.columns = ['Atleta','Potência']
user_power = user_power.groupby('Atleta')['Potência'].mean().reset_index()

fig, ax = plt.subplots(figsize=(6,12))
sns.barplot(data=user_power.sort_values('Potência', ascending=False), y='Atleta', x='Potência', ax=ax, palette='hot')
ax.set_title('Potência Média - 2021')
```
